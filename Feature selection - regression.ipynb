```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv(" .csv")
```

```python
datas=dataset
```

```python
datas=pd.get_dummies(datas,drop_first=True)
```

```python
datas
```

```python
independent=datas.drop('classification_yes',1)
```

```python
dependent=datas['classification_yes']
```

```python
def selectkbest(independent,dependent,n):
    test=SelectKBest(score_func=chi2,k=n)
    test2=test.fit(independent,dependent)
    selectfeature=test2.transform(independent)
    return selectfeature
    
```

```python
def standscaler(independent,dependent):
    X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,y_train,y_test
```

```python
def r2score(regressor,X_test,y_test):
    y_pred=regressor.predict(X_test)
    from sklearn.metrics import r2_score
    r2=r2_score(y_test,y_pred)
    return r2
```

```python
def linear(X_train,y_train,X_test):       
        # Fitting K-NN to the Training set
        from sklearn.linear_model import LinearRegression
        regressor = LinearRegression()
        regressor.fit(X_train, y_train)
        r2=r2score(regressor,X_test,y_test)
        return  r2 
```

```python
def random(X_train,y_train,X_test):
    from sklearn.ensemble import RandomForestRegressor
    RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train,y_train)
    r2=r2score(regressor,X_test,y_test)
    return r2
```

```python
def decision(X_train,y_train,X_test):
    from sklearn.tree import DecisionTreeRegressor
    regressor = DecisionTreeRegressor(random_state = 0)
    regressor.fit(X_train,y_train)
    r2=r2score(regressor,X_test,y_test)
    return r2
```

```python
def select_regression(acclin,accran,accdeci):
    table=pd.DataFrame(index=['chi2score'],columns=['acclin','accran','accdeci'])
    for number,indx in enumerate(table.index):
        table['acclin'][indx]=acclin[number]
        table['accran'][indx]=accran[number]
        table['accdeci'][indx]=accdeci[number]
    return table
```

```python
kbest=selectkbest(independent,dependent,5)
```

```python
acclin=[]
accran=[]
accdeci=[]
```

```python
X_train,X_test,y_train,y_test=standscaler(kbest,independent)
for i in kbest:
    r2_lin=linear(X_train,y_train,X_test)
    acclin.append(r2_lin)
    r2_ran=linear(X_train,y_train,X_test)
    accran.append(r2_ran)
    r2_deci=linear(X_train,y_train,X_test)
    accdeci.append(r2_deci)
   
```

```python
result=select_regression(acclin,accran,accdeci)
```
