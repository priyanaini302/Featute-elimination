```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
```

```python
dataset=pd.read_csv('prep.csv')
```

```python
datas=pd.get_dummies(dataset,drop_first=True)
```

```python
x=datas.drop('classification_yes',1)
```

```python
y=datas['classification_yes']
```

```python
def split_scaler(x,y):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,y_train,y_test
```

```python
def pca():
    from sklearn.decomposition import PCA
    pca=PCA(n_components = 5)
    X_train,X_test,y_train,y_test=split_scaler(x,y)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_
    return  X_train,X_test
```

```python
def select_regression(acclin,accran,accdeci):
    table=pd.DataFrame(index=['chi2score'],columns=['acclin','accran','accdeci'])
    for number,indx in enumerate(table.index):
        table['acclin'][indx]=acclin[number]
        table['accran'][indx]=accran[number]
        table['accdeci'][indx]=accdeci[number]
    return table
```
