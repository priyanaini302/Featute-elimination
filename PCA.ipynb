```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
```

```python
dataset=pd.read_csv('prep.csv')
```

```python
datas=pd.get_dummies(dataset,drop_first=True)
```

```python
datas
```

```python
x=datas.drop('classification_yes',1)
```

```python
y=datas['classification_yes']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
x_test=sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 7)
ind = pca.fit_transform(x)
xtest = pca.transform(x_test)
explained_variance = pca.explained_variance_ratio_
```

```python
acclo=[]
accknn=[]
accran=[]
accdeci=[]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
classifier_lin= LogisticRegression()
classifier_lin.fit(ind, y)
y_pred=classifier_lin.predict(xtest)
acc_lo=accuracy_score(y_test,y_pred)
acclo.append(acc_lo)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
classifier_knn= KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)  
classifier_knn.fit(ind,y)
y_pred=classifier_knn.predict(xtest)
acc_knn=accuracy_score(y_test,y_pred)
accknn.append(acc_knn)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_ran=RandomForestClassifier(n_estimators = 10,criterion = 'entropy', random_state = 0)
classifier_ran.fit(ind,y)
y_pred=classifier_ran.predict(xtest)
acc_ran=accuracy_score(y_test,y_pred)
accran.append(acc_ran)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_deci=DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier_deci.fit(ind,y)
y_pred=classifier_deci.predict(xtest)
acc_deci=accuracy_score(y_test,y_pred)
accdeci.append(acc_deci)
```

```python
 def select_regression(acclo,accknn,accran,accdeci):
    table=pd.DataFrame(index=['pca'],columns=['acclo','accknn','accran','accdeci'])
    for number,indx in enumerate(table.index):
        table['acclo'][indx]=acclo[number]
        table['accknn'][indx]=accknn[number]
        table['accran'][indx]=accran[number]
        table['accdeci'][indx]=accdeci[number]
        
        
    return table
```

```python
result=select_regression(acclo,accknn,accran,accdeci)
```
