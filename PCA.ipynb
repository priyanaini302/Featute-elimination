```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
```

```python
dataset=pd.read_csv('prep.csv')
```

```python
datas=pd.get_dummies(dataset,drop_first=True)
```

```python
datas
```

```python
x=datas.drop('classification_yes',1)
```

```python
y=datas['classification_yes']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
x_test=sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components = 2)
i_ind = pca.fit_transform(x)
x_value = pca.transform(x_test)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
acclin=[]
accran=[]
accdeci=[]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
regressor_lin= LinearRegression()
regressor_lin.fit(i_ind , y)
y_pred=regressor_lin.predict(x_value)
r2_lin=r2_score(y_test,y_pred)
acclin.append(r2_lin)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
regressor_deci = DecisionTreeRegressor(random_state = 0)
regressor_deci.fit(i_ind , y)
y_pred=regressor_deci.predict(x_value)
r2_deci=r2_score(y_test,y_pred)
accdeci.append(r2_deci)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
regressor_ran=RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor_ran.fit(i_ind , y)
y_pred=regressor_ran.predict(x_value)
r2_ran=r2_score(y_test,y_pred)
accran.append(r2_ran)
```

```python
def select_regression(acclin,accdeci,accran):
    table=pd.DataFrame(index=['pca'],columns=['acclin','accran','accdeci'])
    for number,indx in enumerate(table.index):
        table['acclin'][indx]=acclin[number]
        table['accdeci'][indx]=accdeci[number]
        table['accran'][indx]=accran[number]
        
    return table
```

```python
result=select_regression(acclin,accdeci,accran)
```
