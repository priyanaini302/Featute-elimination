```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
```

```python
dataset=pd.read_csv("prep.csv")
```

```python
datas=pd.get_dummies(dataset,drop_first=True)
```

```python
datas
```

```python
independent=datas.drop('classification_yes',1)
dependent=datas['classification_yes']
```

```python
def split_scaler(independent,dependent):
    X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
    sc=StandardScaler()
    X_train=sc.fit_transform(X_train)
    X_test=sc.transform(X_test)
    return X_train,X_test,y_train,y_test
```

```python
def re_ped(regressor,X_test,y_test):
    y_pred=regressor.predict(X_test)
    from sklearn.metrics import r2_score
    r2=r2_score(y_test,y_pred)
    return r2
```

```python
def linar(X_train,y_train,X_test): 
    from sklearn.linear_model import LinearRegression
    regressor=LinearRegression()
    regressor.fit(X_train,y_train)
    r2=re_ped(regressor,X_test,y_test)
    return r2             
```

```python
def decision(X_train,y_train,X_test): 
    from sklearn.tree import DecisionTreeRegressor
    regressor=DecisionTreeRegressor(random_state = 0)
    regressor.fit(X_train,y_train)
    r2=re_ped(regressor,X_test,y_test)
    return r2  
```

```python
def random(X_train,y_train,X_test): 
    from sklearn.ensemble import RandomForestRegressor
    regressor=RandomForestRegressor(n_estimators = 10, random_state = 0)
    regressor.fit(X_train,y_train)
    r2=re_ped(regressor,X_test,y_test)
    return r2  
```

```python
def rfe_sele(independent,dependent,n):
    rfelist=[]
    from sklearn.linear_model import LinearRegression
    lin=LinearRegression()
    from sklearn.tree import DecisionTreeRegressor
    tree=DecisionTreeRegressor(random_state = 0)
    from sklearn.ensemble import RandomForestRegressor
    ran=RandomForestRegressor(n_estimators = 10, random_state = 0)
    lists=[lin,tree,ran]
    for i in lists:
        test=RFE(i,n)
        test1=test.fit(independent,dependent)
        rfe_selec=test1.transform(independent)
        rfelist.append(rfe_selec)
    return rfelist 
```

```python
def dataframe(acclin,accdeci,accran):
    table=pd.DataFrame(index=['linear','decision','random'],columns=['linear','decision','random'])
    for number,indx in enumerate(table.index):
        table['linear'][indx]=acclin[number]
        table['decision'][indx]=accdeci[number]
        table['random'][indx]=accran[number]
    return table
```

```python
acclin=[]
accdeci=[]
accran=[]
```

```python
rfelist=rfe_sele(independent,dependent,3)
```

```python
for i in rfelist:
    X_train,X_test,y_train,y_test=split_scaler(independent,dependent)
    r2_lin=linar(X_train,y_train,X_test)
    acclin.append(r2_lin)
    r2_deci=decision(X_train,y_train,X_test)
    accdeci.append(r2_deci)
    r2_ran=random(X_train,y_train,X_test)
    accran.append(r2_ran)
    
    
```

```python
result=dataframe(acclin,accdeci,accran)
```
